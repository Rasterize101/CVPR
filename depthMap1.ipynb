```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

imgL = cv.imread(r'C:\Users\jitbh\OneDrive\Desktop\DSC_0714.jpg',0)
imgR = cv.imread(r'C:\Users\jitbh\OneDrive\Desktop\DSC_0715.jpg',0)
stereo = cv.StereoBM_create(numDisparities=-6, blockSize=5)
# For each pixel algorithm will find the best disparity from -6
# Larger block size implies smoother, though less accurate disparity map
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity)
plt.colorbar()
plt.show()

win_size = 2
min_disp = -7
max_disp = 9
num_disp = max_disp - min_disp  # Needs to be divisible by 16
stereo = cv.StereoSGBM_create(
    minDisparity=min_disp,
    numDisparities=num_disp,
    blockSize=5,
    uniquenessRatio=5,
    speckleWindowSize=5,
    speckleRange=5,
    disp12MaxDiff=2,
    P1=8 * 3 * win_size ** 2,
    P2=32 * 3 * win_size ** 2,
)
disparity_SGBM = stereo.compute(imgL, imgR)
plt.imshow(disparity_SGBM)
plt.colorbar()
plt.show()
```
